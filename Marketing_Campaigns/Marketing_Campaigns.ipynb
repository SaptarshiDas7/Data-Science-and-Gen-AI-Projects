### Module import

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Step1

### Reading data and trimming column names

```python
df = pd.read_csv("marketing_data.csv")
df.columns = df.columns.str.strip()
df.head()
```

```python
df.shape
```

```python
### Data types of each column
df.dtypes
```

```python
df.describe()
```

```python
df.info()
```

```python
### cleaning 'Dt_Customer' and 'Income' columns
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], format='mixed')

df['Income'] = df['Income'].str.replace('$', '').str.replace(',', '').astype('float64')
df.head()
```

## Step2

```python
df.isnull().sum()
```

```python
### Imputing missing values in 'Income' based on average of Income in that particular group of'Education' and 'Marital_Status'

df['Income'] = df['Income'].fillna(
    df.groupby(['Education', 'Marital_Status'])['Income']
      .transform('mean')
)
df.head(138)
```

## Step 3

```python
### calculating total children
df['Total_Children'] = df['Kidhome'] + df['Teenhome']

### calculating age of customers
year = pd.Timestamp.now().year
df['Age'] =  year - df['Year_Birth']

### calculating total spending
df['Total_Spending'] = df[['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']].sum(axis=1)

### calculating total purchases across three channels
df['Total_Purchases'] = df[['NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases']].sum(axis=1)
df.head()
```

## Step4

### Plotting histograms and boxplots and handling outliers

```python
num_cols = ['Income', 'Total_Children', 'Age', 'Total_Spending', 'Total_Purchases', 'MntWines', 
               'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds', 'NumDealsPurchases', 
               'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases', 'NumWebVisitsMonth']
len(num_cols)
```

```python
### histogram plots for numerical columns

plt.figure(figsize=(20,20))
for i in range(len(num_cols)):
    plt.subplot(6,3,i+1)
    sns.histplot(df[num_cols[i]], kde=True)
    plt.ylabel('')
plt.show()
```

```python
### boxplot for numerical columns

plt.figure(figsize=(30,40))
for i in range(len(num_cols)):
    plt.subplot(6,3,i+1)
    sns.boxplot(df[num_cols[i]])
plt.show()
```

```python
### Most of the numerical columns are right skewed, with some pretty big values.
```

```python
### Seeing plots of income based on Education and Marital_Status
plt.figure(figsize=(18,10))
sns.boxplot(y = 'Income', x = 'Education', data=df, palette='Set2',hue='Marital_Status')
plt.show()
```

```python
### outlier handling using winsorization
from scipy.stats.mstats import winsorize
def winsorize_group(x, lower=0.00, upper=0.05):      ### clipping top 5% values considering the boxplot
    return winsorize(x, limits=[lower, upper])

df['Income_wins'] = (
    df
    .groupby(['Education', 'Marital_Status'])['Income']
    .transform(lambda x: winsorize_group(x))
)
df[['Income', 'Income_wins']].describe()
```

```python
### boxplot after winsorization
plt.figure(figsize=(18,10))
sns.boxplot(y = 'Income_wins', x = 'Education', data=df, palette='Set2',hue='Marital_Status')
plt.show()
```

```python
### The top 5% outliers have been successfully handled using winsorization technique.
### Now the boxplots look much better and more interpretable.
```

## Step7

### Hypothesis Testing

```python
### a.	Older individuals may not possess the same level of technological proficiency and may, therefore, lean toward traditional in-store shopping preferences.
from scipy import stats

age_above_60 = df[df['Age'] >= 60]
t_score, p_value = stats.ttest_rel(age_above_60['NumStorePurchases'], age_above_60['NumWebPurchases'], alternative='greater')
print(f"T-score: {t_score}, P-value: {p_value}")
if p_value < 0.05:
    print("Reject the null hypothesis: Older customers lean toward traditional in-store shopping preferences.")
else:
    print("Fail to reject the null hypothesis: Older customers do not lean toward traditional in-store shopping preferences.")

```

```python
age_above_60['NumStorePurchases'].mean(), age_above_60['NumWebPurchases'].mean()
### Store purchases mean for 60+ people is more, thereby validating our hypothesis test claim.
```

```python
### b.	Customers with children likely experience time constraints, making online shopping a more convenient option.
from scipy import stats

children = df[df['Total_Children'] > 0]

t_score, p_value = stats.ttest_rel(
    children['NumWebPurchases'],
    children['NumStorePurchases'],
    alternative='greater'
)

print(f"T-score: {t_score}, P-value: {p_value}")

if p_value < 0.05:
    print("Reject the null hypothesis: Customers with children show a statistically significant preference for online shopping.")
else:
    print("Fail to reject the null hypothesis: No statistical evidence that customers with children prefer online shopping.")

```

```python
children['NumWebPurchases'].mean(), children['NumStorePurchases'].mean()
### Web Purchases mean is lower validating our hypothesis test claim.
```

```python
### c.	Sales at physical stores may face the risk of cannibalization by alternative distribution channels.
from scipy import stats

df['Other_Purchase_Mode'] = df['NumWebPurchases'] + df['NumCatalogPurchases']

t_score, p_value = stats.ttest_rel(
    df['NumStorePurchases'],
    df['Other_Purchase_Mode'],
    alternative='less'
)

print(f"T-score: {t_score}, P-value: {p_value}")

if p_value < 0.05:
    print("Reject the null hypothesis: Sales at physical stores may face the risk of cannibalization by alternative distribution channels.")
else:
    print("Fail to reject the null hypothesis: No statistical evidence of cannibalization.")
```

```python
df['NumStorePurchases'].mean(),df['Other_Purchase_Mode'].mean()
### StorePurchases mean is less, thereby validating our hypothesis test result.
```

```python
### d.	Does the United States significantly outperform the rest of the world in total purchase volumes?
from scipy import stats

df_US = df_enc.loc[df_enc['Country_US'] == 1]
df_non_US = df_enc.loc[df_enc['Country_US'] == 0]
t_score, p_value = stats.ttest_ind(df_US['Total_Purchases'], df_non_US['Total_Purchases'], alternative='greater')
print(f"T-score: {t_score}, P-value: {p_value}")
if p_value < 0.05:
    print("Reject the null hypothesis: US customers significantly outperform the rest of the world in total purchase volumes.")
else:
    print("Fail to reject the null hypothesis: No evidence that US customers significantly outperform non-US customers.")
    
```

```python
df_US['Total_Purchases'].mean(), df_non_US['Total_Purchases'].mean()  
### US purchases has higher mean, validating our hypothesis test result.
```

```python
### Inferences from Hypothesis Testing:
# a. Older customers tend to prefer in-store shopping over online shopping, as indicated by the significant difference in purchase modes.
# b. No statistical evidence that customers with children prefer online shopping.
# c. Sales at physical stores may face the risk of cannibalization by alternative distribution channels.
# d. US customers significantly outperform the rest of the world in total purchase volumes."

### We have performed ttest_ind for q)d because the two samples are independent(US and Non-US). However, in the other 3 questions, we are testing
### on different traits of the same sample and hence performing ttest_rel.

### We have performed one tailed t-test in each of the questions as we wanted to know one direction greater/lesser and not just same or different.
```

```python
### b.	Examine if there is a correlation between customers' age and the acceptance rate of the last campaign.

### Violin plot to check distribution and spread

plt.figure(figsize=(10,6))
sns.violinplot(x='Response', y='Age', data=df)
plt.title('Age vs Acceptance of Last Campaign')
plt.xlabel('Accepted Last Campaign (0=No, 1=Yes)')
plt.ylabel('Age')
plt.show()
```

```python
### Acceptance rate in various age groups

df['Age_Group'] = pd.cut(
    df['Age'],
    bins=[0,30,40,50,60,100],
    labels=['<30','30–40','40–50','50–60','60+']
)

acc_rate = df.groupby('Age_Group')['Response'].mean()

acc_rate.plot(kind='bar', figsize=(8,5))
plt.title('Campaign Acceptance Rate by Age Group')
plt.ylabel('Acceptance Rate')
plt.show()
```

```python
### Point biserial correlation between Age and Response

from scipy.stats import pointbiserialr

# Drop missing values if any
df_pb = df[['Age', 'Response']].dropna()

corr, p_value = pointbiserialr(
    df_pb['Response'],
    df_pb['Age']
)

print(f"Point-biserial correlation: {corr:.3f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("There is a significant correlation between age and acceptance of the last campaign.")
else:
    print("There is no significant correlation between age and acceptance of the last campaign.")
```

```python
### Inference:
# Although younger customers appear to have marginally higher campaign acceptance rates in the visual analysis, the age distributions of responders and non-responders largely overlap. 
# This is supported by a non-significant point-biserial correlation (r = −0.021, p = 0.313), indicating that customer age does not significantly influence acceptance of the last campaign.
```
