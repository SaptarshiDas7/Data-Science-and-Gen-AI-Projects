```python
from datetime import datetime
import json
import hashlib

tasks={}
user_name=''

# Function to hash the password in jason file for security purposes
def hash_password(password):
    """Hash a password using SHA-256."""
    return hashlib.sha256(password.encode()).hexdigest()

# User authentication
def authentication():
    """
    Authenticate a user by username and password.
    
    The function prompts for username and password. If the user doesn't exist,
    creates a new user with the provided credentials. If the user exists,
    validates the password and allows login on correct password.
    
    Global variables modified:
        tasks (dict): The tasks dictionary containing user data
        user_name (str): The current user's username
    """
    global tasks, user_name
    user_name = input('Enter your user name')
    password = input('Enter your password')
    if user_name not in tasks:
        tasks[user_name]={}
        tasks[user_name]['password'] = hash_password(password)
    else:
        while True:
            if tasks[user_name]['password'] == hash_password(password):
                print('Logged in')
                break
            else:
                print('Password is incorrect! Try again...')
                password = input('Enter your password')
                continue

# Function to add a task
def add_task():
    """
    Add a new task for the current user.
    
    This function adds a task with a description and status to the user's task list.
    Tasks are stored with auto-generated IDs in the format 'Task1', 'Task2', etc.
    
    Global variables:
        tasks (dict): Dictionary storing all users' tasks
        user_name (str): Current logged in username
        
    Returns:
        None
    """
    global tasks, user_name
    id = len(tasks[user_name])
    desc = input('Enter task description: ')
    status = input('Enter task status as Completed/Pending: ')
    while status.lower() not in ('completed','pending'):
        print('Enter correct value in Status')
        status = input('Enter task status as Completed/Pending: ')
        continue
    tasks[user_name]['Task'+str(id)]={'description' : desc, 'status' :status.lower()}

# Function to view all the tasks
def view_tasks():
    """
    Display all tasks for the currently authenticated user.

    Prints the user name and each task's ID, description, and status.
    The special 'password' key is skipped when listing tasks.

    Globals:
        tasks (dict): Dictionary storing all users' tasks.
        user_name (str): Currently authenticated username.
    """
    global tasks, user_name
    print(f'\n User Name: {user_name}')
    for key in tasks[user_name]:
        if key!='password':
            print(f'ID: {key}      Description: {tasks[user_name][key]['description']}     Idea: {tasks[user_name][key]['status']}\n')


# Function to change the status of a task 
def change_status():
    """
    Mark a specified task as completed for the current user.

    Prompts for a Task ID (e.g., "Task1") and, if the ID exists for the user,
    updates the task's 'status' field to 'Completed'. Prints an error message
    if the ID is invalid.

    Globals:
        tasks (dict): Dictionary storing all users' tasks.
        user_name (str): Currently authenticated username.
    """
    global tasks, user_name
    id = input('Enter Task ID as Task1,Task2....: ')
    if id not in tasks[user_name]:
        print('Invalid ID!')
    else:
        for key in tasks[user_name]:
            if key == id:
                tasks[user_name][key]['status'] = 'Completed'

# Function to delete a task
def delete_task():
    """
    Delete a task for the current user and reindex remaining tasks.
    
    Prompts user for a Task ID and removes it if valid. After deletion,
    remaining tasks are reindexed to maintain sequential Task IDs.
    Preserves the user's password during reindexing.
    
    Global variables:
        tasks (dict): Dictionary storing all users' tasks
        user_name (str): Current logged in username
    """
    global tasks, user_name
    id = input('Enter Task ID as Task1,Task2....: ')
    if id not in tasks[user_name]:
        print('Invalid ID!')
    else:
        del tasks[user_name][id]
        temp = {}
        i = 1
        for key, value in tasks[user_name].items():
            if key != 'password':  # Don't touch the password key
                temp[f'Task{i}'] = value
                i += 1
        
        # Preserve password and update the user's data
        password = tasks[user_name]['password']
        tasks[user_name] = {'password': password}
        tasks[user_name].update(temp)


# Function to save the tasks
def save_tasks():
    """
    Save all tasks to a JSON file.
    
    Writes the tasks dictionary to 'tasks.json' file in JSON format.
    
    Global variables:
        tasks (dict): Dictionary storing all users' tasks
        user_name (str): Current logged in username
    """
    global tasks, user_name
    with open('tasks.json','w') as file:
        json.dump(tasks,file)
    print('Tasks saved successfully')
    
# Function to load the tasks
def load_tasks():
    """
    Load tasks from the JSON file.
    
    Attempts to read tasks from 'tasks.json' file. If file doesn't exist,
    prints an error message.
    
    Global variables:
        tasks (dict): Dictionary storing all users' tasks
        user_name (str): Current logged in username
    """
    global tasks, user_name
    try:
        with open('tasks.json','r') as file:
            tasks=json.load(file)
        print('Loaded successfully')
    except FileNotFoundError:
        print('No file found')


        
# Function to display the menu
def menu():
    """
    Display and handle the main menu interface.
    
    Loads tasks, authenticates user, and provides a menu-driven interface
    for task management operations including add, view, change status,
    delete, and logout options. Saves tasks before logout.
    """
    load_tasks()
    print(tasks,type(tasks))
    authentication()
    
    while True:
        print("\nPersonal Expense Tracker")
        print("1. Add Task")
        print("2. View Tasks")
        print("3. Change Status")
        print("4. Delete task")
        print("5. Logout")

        choice = input("Select an option: ")

        if choice == '1':
            add_task()
        elif choice == '2':
            view_tasks()
        elif choice == '3':
            change_status()
        elif choice == '4':
            delete_task()     
        elif choice == '5':
            save_tasks()  # Save expenses before exiting
            print("Logged out")
            break
        else:
            print("Invalid choice. Please try again.")

while True:    
    opt = input('Enter Y to search for another user and N to stop.')
    if opt.strip() not in ('Y','N'):
        print('Invalid input!!!')
        continue
    if opt.strip()=='Y':
        menu()
        continue
    else:
        print('Thanks for using the task manager!')
        break
    
```
