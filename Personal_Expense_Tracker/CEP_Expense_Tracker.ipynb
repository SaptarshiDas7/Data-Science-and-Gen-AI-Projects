Objectives:

1.
Design and implement a personal expense tracker that enables users to manage their expenses

2.
Allow users to categorize expenses and set monthly budgets

3.
Implement file-handling functionality to save and load expense data

4.
Create an interactive, menu-driven interface for ease of use

```python
### list to hold expenses
from datetime import datetime
expenses=[]
# monthly_budget=0
import json

# function to add expenses
def add_expenses():
    """Prompt the user to enter a new expense and add it to the global expenses list.

    Prompts for:
    - date in YYYY-MM-DD format (validated)
    - category (string)
    - amount (non-negative float, validated)
    - description (string)

    On success, appends a dict with keys 'date' (ISO string), 'category', 'amount', and
    'description' to the global `expenses` list and sorts the list by date. Prints a
    confirmation message. Handles invalid input by re-prompting.
    """
    while True:
        date = input('Enter the date of expense (YYYY-MM-DD): ').strip()
        try:
            date_obj = datetime.strptime(date, "%Y-%m-%d").date()
            break
        except ValueError:
            print('Enter date in correct format (YYYY-MM-DD).')
    category = input('Enter the category of the expense: ').strip()
    while True:
        try:
            amount = float(input('Enter the amount spent: ').strip())
            if amount < 0:
                print("Don't enter a negative value for amount.")
                continue
            break
        except ValueError:
            print('Enter amount in a valid numeric format.')
    description = input('Enter a brief description of the expense: ').strip()

    expense = {
        'date': date_obj.isoformat(),
        'category': category,
        'amount': amount,
        'description': description
    }

    expenses.append(expense)
    expenses.sort(key=lambda x: datetime.strptime(x['date'], "%Y-%m-%d"))
    print('Expense added successfully.')

# function to view expenses
def view_expenses():
    """Print all stored expenses from the global `expenses` list.

    If no expenses are stored, prints a message and returns. For each expense,
    verifies presence of the keys 'date', 'category', 'amount', and 'description'
    before printing; incomplete entries are skipped with a notice.
    """
    if not expenses:
        print('\nNo stored expenses.')
        return
    print('\nStored expenses:')
    for expense in expenses:
        if all(k in expense for k in ('date', 'category', 'amount', 'description')):
            print(f"Date: {expense['date']}, Category: {expense['category']}, Amount: ${float(expense['amount']):.2f}, Description: {expense['description']}")
        else:
            print('Incomplete entry found, skipped.')

# function to track the budget of a particular month
def track_budget():
    """Prompt for a month (MM-YYYY) and monthly budget, then report totals.

    Validates month format MM-YYYY and budget as a non-negative float. Computes
    the sum of expenses in the specified month/year from the global `expenses`
    list and prints the total and whether the budget was exceeded, including
    remaining balance if not exceeded.
    """
    while True:
        month = input('Enter month to track (MM-YYYY): ').strip()
        try:
            month_obj = datetime.strptime(month, "%m-%Y")
            break
        except ValueError:
            print('Enter month in correct format (MM-YYYY).')
    while True:
        try:
            monthly_budget = float(input('Enter your monthly budget: ').strip())
            if monthly_budget < 0:
                print("Don't enter a negative value for monthly budget.")
                continue
            break
        except ValueError:
            print('Enter the monthly budget in a valid numeric format.')

    total_expenses = sum(
        float(exp['amount']) for exp in expenses
        if datetime.strptime(exp['date'], "%Y-%m-%d").month == month_obj.month
        and datetime.strptime(exp['date'], "%Y-%m-%d").year == month_obj.year
    )

    print(f"Total expenses for {month_obj.strftime('%m-%Y')}: ${total_expenses:.2f}")
    if total_expenses > monthly_budget:
        print('You have exceeded your budget.')
    else:
        remaining_balance = monthly_budget - total_expenses
        print(f'You have ${remaining_balance:.2f} left for the month.')

# function to save expenses to a file
def save_expenses():
    """Save the global `expenses` list to 'expenses.json'.

    Uses json.dump with indentation. Prints success or an error message if an
    exception occurs during file write.
    """
    try:
        with open('expenses.json', 'w') as file:
            json.dump(expenses, file, indent=2)
        print('Expenses saved successfully.')
    except Exception as e:
        print(f'Error saving expenses: {e}')

# function to retrieve the expenses records
def load_expenses():
    """Load expenses from 'expenses.json' into the global `expenses` list.

    If the file does not exist or is invalid/corrupted, initializes `expenses`
    as an empty list and prints a notice. If data is a list, replaces the global
    `expenses` with it and sorts by date.
    """
    global expenses
    try:
        with open('expenses.json', 'r') as file:
            data = json.load(file)
            if isinstance(data, list):
                expenses = data
                expenses.sort(key=lambda x: datetime.strptime(x['date'], "%Y-%m-%d"))
                print('Expenses loaded successfully.')
            else:
                print('Data in file is not a list. Starting with empty expenses.')
                expenses = []
    except FileNotFoundError:
        expenses = []
        print('No expenses file found; starting fresh.')
    except json.JSONDecodeError:
        expenses = []
        print('Expenses file is corrupted; starting with empty expenses.')
        
# Function to display the menu
def menu():
    load_expenses()  # Load expenses when the program starts
    while True:
        print("\nPersonal Expense Tracker")
        print("1. Add Expense")
        print("2. View Expenses")
        print("3. Track Budget")
        print("4. Save Expenses")
        print("5. Exit")

        choice = input("Select an option: ")

        if choice == '1':
            add_expenses()
        elif choice == '2':
            view_expenses()
        elif choice == '3':
            track_budget()
        elif choice == '4': 
            save_expenses()
        elif choice == '5':
            save_expenses()  # Save expenses before exiting
            print("Exiting the program.")
            break
        else:
            print("Invalid choice. Please try again.")

# Run the menu function
menu()       
```
